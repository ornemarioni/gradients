### Potencial esfera isoterma


```python
import numpy as np
import matplotlib.pyplot as plt

def potencial(R,z=0):
    potencial = np.log(np.sqrt(R**2+z**2))
    return potencial

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,7))
fig.subplots_adjust(bottom=0.1, left =0.12, right = 0.97, top = 0.93, wspace=0., hspace= 0)

R = np.logspace(np.log10(0.2),np.log10(50),300)

phi1=potencial(R)
phi2=potencial(R,z=5)
phi3=potencial(R,z=10)
ax.plot(R,phi1, label='z=0')
ax.plot(R,phi2, label='z=5')
ax.plot(R,phi3, label='z=10')
ax.minorticks_on()
ax.tick_params( labelsize=18)
ax.tick_params('both', length=5, width=1.2,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=8, width=1.2,which='major', direction='in', right='on',top='on')
ax.set_ylabel('$\Phi$',fontsize=18)
ax.set_xlabel('R',fontsize=18)
ax.legend(fontsize=15)
plt.show()
```

### Fuerza en z

```python
import numpy as np
import matplotlib.pyplot as plt

R = (2,4,6,8,10,12,14)

def fuerza(R,z=0):
    aux = (R**2+z**2)**2
    fuerza = z/aux
    return fuerza

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,7))
fig.subplots_adjust(bottom=0.1, left =0.12, right = 0.97, top = 0.93, wspace=0., hspace= 0)

R = np.logspace(np.log10(0.02),np.log10(20),300)

F1=fuerza(R)
F2=fuerza(R,z=5)
F3=fuerza(R,z=10)
ax.plot(R,F1,label='z=0')
ax.plot(R,F2,label='z=5')
ax.plot(R,F3,label='z=10')
ax.minorticks_on()
ax.tick_params( labelsize=18)
ax.tick_params('both', length=5, width=1.2,which='minor', direction='in', right='on',top='on')
ax.tick_params('both', length=8, width=1.2,which='major', direction='in', right='on',top='on')
ax.set_ylabel('aceleracion',fontsize=18)
ax.set_xlabel('R',fontsize=18)
ax.legend(fontsize=15)

plt.show()
```
